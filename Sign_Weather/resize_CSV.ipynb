```python
import numpy
import csv
%cd /content/drive/MyDrive/Sign
import pandas as pd

def resize():
    # 사용할 변수 선언
    n_steps = 1
    time_periods = 20
    number_point = 84
    label_column = number_point + 1

    # CSV파일 load
    raw_dataset = numpy.loadtxt("/content/drive/MyDrive/Sign/CSV/HAND_ALL.csv", delimiter=",")

    # Keypoint 데이터 가져오기 (the coordinates)
    X = raw_dataset[:, 0:number_point]

    # Frame번호 가져오기
    time_stamps = raw_dataset[:, number_point]

    # Get the labels
    Y = raw_dataset[:, label_column]
    lap_buffer = []
    X, Y = create_segments_and_labels(X, time_periods, n_steps, Y)

    # frames을 one row로 변경
    for i in range(len(X)):
        row = X[i] + [Y[i]]
        lap_buffer.append(row)
        print(i)
    len(lap_buffer)
    out_df = pd.DataFrame(lap_buffer)
    out_df.to_csv('/content/drive/MyDrive/Sign/HAND_One.csv',header=False, index=False) #csv파일로 저장 

def create_segments_and_labels(df, time_steps, step, label_name):
    N_FEATURES = 84
    segments = []
    label_array = []
    df = numpy.ndarray.tolist(df)
    for i in range(0, (len(df) - time_steps)+1, time_steps):
        base = df[i]
        #print("base i:")
        #print(i)
        for j in range(1, time_steps):
            base = base + df[i + j]
        segments.append(base)
        label_array.append(label_name[i])

    #print("Segment: ", segments, "\n")

    return segments, label_array

resize()
print('Success!')
```
