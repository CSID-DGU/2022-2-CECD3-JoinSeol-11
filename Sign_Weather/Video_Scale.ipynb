##Data Augmentation - rescale

```python
# 하나의 video에 대한 코드
import cv2
import numpy as np

cap = cv2.VideoCapture('/content/drive/MyDrive/AIHUB/WEATHER/54/NIA_SL_WORD1140_REAL16_F.mp4')

fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter('/content/drive/MyDrive/output.mp4',fourcc, 30, (1920,1440))

while True:
    ret, frame = cap.read()
    if ret == True:
        b = cv2.resize(frame,(1920,1440),fx=0,fy=0, interpolation = cv2.INTER_CUBIC)
        out.write(b)
    else:
        break
    
cap.release()
out.release()
cv2.destroyAllWindows()
```
